```python
import os
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors
import matplotlib.pyplot as plt
import seaborn as sns

base_dir = r"C:\Users\avari"
raw_dir = os.path.join(base_dir, "data", "raw")
processed_dir = os.path.join(base_dir, "data", "processed")
results_dir = os.path.join(base_dir, "data", "results")

csv_path = os.path.join(raw_dir, "coconut_csv_lite-05-2026.csv")
df = pd.read_csv(csv_path, low_memory=False)

subset = df.sample(n=2000, random_state=42).copy()

smiles_col = "canonical_smiles"

subset["mol"] = subset[smiles_col].apply(
    lambda x: Chem.MolFromSmiles(str(x)) if pd.notnull(x) else None
)

clean = subset[subset["mol"].notnull()].copy()

print("Subset size:", len(subset))
print("Valid molecules:", len(clean))

def calc_desc(mol):
    return {
        "MolWt_rdkit": Descriptors.MolWt(mol),
        "LogP_rdkit": Descriptors.MolLogP(mol),
        "TPSA_rdkit": Descriptors.TPSA(mol),
        "HDonors_rdkit": Descriptors.NumHDonors(mol),
        "HAcceptors_rdkit": Descriptors.NumHAcceptors(mol),
        "RotBonds_rdkit": Descriptors.NumRotatableBonds(mol),
    }

desc_df = clean["mol"].apply(calc_desc).apply(pd.Series)

clean = pd.concat([clean.reset_index(drop=True), desc_df.reset_index(drop=True)], axis=1)

clean_path = os.path.join(processed_dir, "coconut_subset_cleaned.csv")
clean.to_csv(clean_path, index=False)
clean_path
```

```python
import os
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors
import matplotlib.pyplot as plt
import seaborn as sns

base_dir = r"C:\Users\avari"
raw_dir = os.path.join(base_dir, "data", "raw")
processed_dir = os.path.join(base_dir, "data", "processed")
results_dir = os.path.join(base_dir, "data", "results")

os.makedirs(processed_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

csv_path = os.path.join(raw_dir, "coconut_csv_lite-05-2026.csv")
df = pd.read_csv(csv_path, low_memory=False)

subset = df.sample(n=2000, random_state=42).copy()

smiles_col = "canonical_smiles"

subset["mol"] = subset[smiles_col].apply(
    lambda x: Chem.MolFromSmiles(str(x)) if pd.notnull(x) else None
)

clean = subset[subset["mol"].notnull()].copy()

print("Subset size:", len(subset))
print("Valid molecules:", len(clean))

def calc_desc(mol):
    return {
        "MolWt_rdkit": Descriptors.MolWt(mol),
        "LogP_rdkit": Descriptors.MolLogP(mol),
        "TPSA_rdkit": Descriptors.TPSA(mol),
        "HDonors_rdkit": Descriptors.NumHDonors(mol),
        "HAcceptors_rdkit": Descriptors.NumHAcceptors(mol),
        "RotBonds_rdkit": Descriptors.NumRotatableBonds(mol),
    }

desc_df = clean["mol"].apply(calc_desc).apply(pd.Series)

clean = pd.concat([clean.reset_index(drop=True), desc_df.reset_index(drop=True)], axis=1)

clean_path = os.path.join(processed_dir, "coconut_subset_cleaned.csv")
clean.to_csv(clean_path, index=False)
clean_path
```

```python
desc_cols = [
    "molecular_weight",
    "alogp",
    "topological_polar_surface_area",
    "hydrogen_bond_donors",
    "hydrogen_bond_acceptors",
    "MolWt_rdkit",
    "LogP_rdkit",
    "TPSA_rdkit",
    "HDonors_rdkit",
    "HAcceptors_rdkit",
    "RotBonds_rdkit",
]

descriptor_table = clean[desc_cols].copy()
desc_path = os.path.join(processed_dir, "descriptor_table.csv")
descriptor_table.to_csv(desc_path, index=False)
desc_path
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.histplot(clean["MolWt_rdkit"], bins=40)
plt.xlabel("Molecular weight (RDKit)")
plt.title("Natural-product molecular weight distribution")
plt.tight_layout()
plt.savefig(os.path.join(results_dir, "molwt_hist.png"), dpi=300)
plt.close()

plt.figure(figsize=(6,4))
sns.histplot(clean["LogP_rdkit"], bins=40)
plt.xlabel("LogP (RDKit)")
plt.title("Natural-product LogP distribution")
plt.tight_layout()
plt.savefig(os.path.join(results_dir, "logp_hist.png"), dpi=300)
plt.close()

plt.figure(figsize=(6,4))
sns.scatterplot(data=clean, x="MolWt_rdkit", y="LogP_rdkit", alpha=0.5)
plt.xlabel("Molecular weight (RDKit)")
plt.ylabel("LogP (RDKit)")
plt.title("MW vs LogP for COCONUT subset")
plt.tight_layout()
plt.savefig(os.path.join(results_dir, "mw_vs_logp.png"), dpi=300)
plt.close()

plt.figure(figsize=(6,5))
corr_cols = ["MolWt_rdkit", "LogP_rdkit", "TPSA_rdkit", "HDonors_rdkit", "HAcceptors_rdkit", "RotBonds_rdkit"]
corr = clean[corr_cols].corr()
sns.heatmap(corr, annot=True, cmap="viridis")
plt.title("Descriptor correlation heatmap")
plt.tight_layout()
plt.savefig(os.path.join(results_dir, "descriptor_heatmap.png"), dpi=300)
plt.close()
```
